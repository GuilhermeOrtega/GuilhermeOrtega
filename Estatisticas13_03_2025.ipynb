DetecÃ§Ã£o

* Contagem das linhas


```python
df.isna().sum(axis = 1)
```

```python
#Axis = 1 pega as linhas e soma, = 0 as colunas
#Criando um missing por linhas
missings_por_linha = pd.DataFrame({'id': df['id'].to_list(),
              'n_missings': df.isna().sum(axis = 1)})
```

```python
missings_por_linha
```

```python
n_colunas = df.shape[1]-1
n_colunas
```

```python
#Porcentual de quanto cada linha peder com missing
missings_por_linha['perc_missing'] = missings_por_linha['n_missings'] / n_colunas
```

```python
missings_por_linha.sort_values('perc_missing', ascending = False)
```

```python
#Filtrar do original
ids_problema = missings_por_linha[missings_por_linha['perc_missing'] > 0.2]['id'].values
ids_problema
```

```python
df_filtrado = df[~df['id'].isin(ids_problema)]
```

```python
df[~df['id'].isin(ids_problema)]
```

```python
df.shape, df_filtado.shape
```

* Contegem de Colunas

```python
contagem_colunas = df.isna().sum(axis = 0).reset_index()
```

```python
contagem_colunas  = contagem_colunas.rename(columns = {'index':'colunas', 0:'n_missings'})
```

```python
n_linhas = df.shape[0]
n_linhas
```

```python
contagem_colunas
```

```python
contagem_colunas['perc_missings'] = contagem_colunas['n_missings'] / n_linhas
```

```python
contagem_colunas
```

```python
contagem_colunas.sort_values('perc_missings', ascending = False)
```

#IndentificaÃ§Ã£o

* MÃ©todo Score Z

```python
# idade
dados_credito['idade_z'] = (dados_credito['idade'] - np.mean(dados_credito['idade'])) / np.std(dados_credito['idade'])
```

```python
#Se nÃ£o houver retorno Ã© pq a coluna nÃ£o tem outliners
dados_credito[~dados_credito['idade_z'].between(-3,3)]
```

```python
# renda
dados_credito['renda_z'] = (dados_credito['renda'] - np.mean(dados_credito['renda'])) / np.std(dados_credito['renda'])
```

```python
dados_credito[~dados_credito['renda_z'].between(-3,3)]
```
